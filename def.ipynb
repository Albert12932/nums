## <font color="35e8b2">Что такое функции?</font>
- Функции - это участки кода, которые выполняют определенную задачу при вызове.
- Они позволяют организовать программу на более удобные и структурированные блоки.

## <font color="35e8b2">Примеры из реальной жизни:</font>
- Функция "стиралка": принимает грязную одежду в качестве аргумента, возвращает чистую одежду.


## <font color="35e8b2">Структура функция в Python:</font>

```python
def название функции(аргументы):
    код
    return результат
```

## <font color="35e8b2">Примеры функций в Python:</font>

### Пример функции без аргументов:


````python
def hello():
    print("Привет, мир!")

hello()
````

````python
def hello():
    return 'Привет, мир!'

hello()
````

### Пример функции с аргументами:

````python
def add_numbers(a, b):
    return a + b

result = add_numbers(2, 3)
print(result)
````

## <font color="35e8b2">Локальные переменные в Python:</font>

- Локальные переменные - объявлены внутри функции и доступны только внутри этой функции.

### <font color="35e8b2">Пример на Python:</font>

````python
def multiply_numbers(a, b):
    result = a * b  # локальная переменная result
    return result

print(multiply_numbers(3, 4))
print(result)  # Ошибка: result не определен вне функции
````

# <font color="35e8b2">Глобальные переменные в Python:</font>
- Глобальные переменные - объявлены вне функции и доступны в любом месте программы.
### <font color="35e8b2">Пример на Python:</font>

````python
cnt = 0  # глобальная переменная

def add(a,b):
    global cnt
    cnt += 1
    return a+b


print(add(5,6))
print(cnt)
````

## Особенности работы с локальными и глобальными переменными:
- Локальные переменные имеют более высокий приоритет по сравнению с глобальными внутри функции.
- Для изменения значения глобальной переменной внутри функции необходимо использовать ключевое слово global.
- Изменения в глобальной переменной, сделанные внутри функции, сохраняются после выполнения функции.
