# 25 номер ЕГЭ. Маски

```python
from fnmatch import fnmatch

```

```python
for x in range(0, 10**8+1, 1991):
    if fnmatch(str(x), '2*1?71'):
        print(x, x//1991)

```

```python
for x in range(0, 10**8, 3123):
    if fnmatch(str(x), '3?1*57'):
        print(x, x//3123)
```

```python
for x in range(0, 10**8, 2024):
    if fnmatch(str(x), '3?2758*4'):
        print(x)
```

```python
for x in range(0, 10**8, 2024):
    if fnmatch(str(x), '1?4*78*') and sum(map(int, str(x))) < 35:
        print(x)
```

```python
for x in range(0, 10**8, 3123):
    if fnmatch(str(x), '3?1*57') and bin(x).count('1') < 11:
        print(x)
```
