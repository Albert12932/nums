### Itertools

```python
from itertools import product, permutations
```

```python
for i in product('123', repeat=3):
    x = ''.join(i)
    print(x)
```

```python
for i in permutations('123'):
    x = ''.join(i)
    print(x)
```

```python
num = 0

for i in product(sorted('КАЛЕЙДСОП', reverse=True), repeat=6):
    x = ''.join(i)
    if (num%2==0) and (x[0] == 'К') and (x.count('Й') >= 2) and ('С' not in x) and ('Е' not in x):
        print(num, x)
    num += 1

# 243698 
```
