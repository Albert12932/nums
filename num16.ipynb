```python
def t23(start, finish):
    if start==finish:
        return 1
    if start>finish:
        return 0
    return t23(start+2, finish)+t23(start+7, finish)

print(t23(7,25) * t23(25, 51))
```

```python
def F(n):
    if n==0:
        return 0
    if 0 < n < 3:
        return 1
    if n >= 3:
        return F(n-2)+F(n-1)

print(F(30))

```

```python
def F(n):
    if n == 0:
        return 1
    if 0 < n <= 10:
        return F(n-1)
    if 10 < n < 100:
        return 2.2*F(n-3)
    if n >= 100:
        return 1.7*F(n-2)

print(F(22))

# 23
```

```python
def F(n):
    if n <= 1:
        return 1
    if n > 1 and n % 2 != 0:
        return 5*n + F(n-1) + F(2)
    if n > 1 and n % 2 == 0:
        return 3*F(n-1)
    
print(F(23))

# 2214271
```

```python
def F(n):
    if n <= 9:
        return n
    if n > 9 and n % 5 == 0:
        return n + F(n//5+2)
    if n > 9 and n % 5 != 0:
        return n + F(n+5)
    
for n in range(1, 1000):
    try:
        if F(n) > 999:
            print(n)
    except RecursionError:
        continue
```

```python
def F(n):
    if n > 55:
        return n*n + 5*n + 5
    if n <= 55 and n % 5 == 0:
        return F(n+1) + F(n+2) + 2*F(n+5)
    if n <= 55 and n % 5 != 0:
        return F(n+3) + 6*F(n+4)

cnt = 0
for n in range(1, 1000):
    if sum(map(int, str(F(n)))) == 19:
        cnt += 1

print(cnt)
```
