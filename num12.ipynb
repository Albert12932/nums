```python
s = '8'*247

while '888' in s or '111' in s:
    if '888' in s:
        s = s.replace('888', '1', 1)
    else:
        s = s.replace('111', '8', 1)

print(s)
```

```python
s = '4' + '6'*79 + '4'

while '63' in s or '664' in s or '6665' in s:
    if '63' in s:
        s = s.replace('63', '4', 1)
    elif '664' in s:
        s = s.replace('664', '5', 1)
    elif '6665' in s:
        s = s.replace('6665', '3', 1)

print(s)
```

```python
s = '>' + '1'*10 + '2'*20 + '3'*30

while '>1' in s or '>2' in s or '>3' in s:
    if '>1' in s:
        s = s.replace('>1', '22>', 1)
    if '>2' in s:
        s = s.replace('>2', '2>', 1)
    if '>3' in s:
        s = s.replace('>3', '1>', 1)

print(sum(map(int, s.replace('>', ''))))
```

```python
s = '1'*25 + '2'*25 + '3'*25 + '1'* 25 + '3'*25 + '2'*25

while '12' in s or '32' in s or '31' in s:
    if '12' in s:
        s = s.replace('12', '21', 1)
    if '32' in s:
        s = s.replace('32', '23', 1)
    if '31' in s:
        s = s.replace('31', '13', 1)

print(s[19]+s[79]+s[119])
print(s)
```

```python
stroki = []

for n in range(2, 100):
    s = '8'*n
    while '555' in s or '888' in s:
        s = s.replace('555', '8', 1)
        s = s.replace('888', '55', 1)
    if s not in stroki:
        stroki.append(s)

print(len(stroki))
```

```python
minimal = 10**10

for n in range(101, 1000):
    s = '1'*n
    while '111' in s:
        s = s.replace('111', '22', 1)
        s = s.replace('222', '11', 1)
    
    if s.count('1') < minimal:
        print(n, s.count('1'))
        minimal = s.count('1')
```
