# 25 номер ЕГЭ. Делители

```python
for x in range(174457, 174506):
    dels = set()
    for delitel in range(2, int(x**0.5)+1):
        if x%delitel == 0:
            dels.add(delitel)
            dels.add(x//delitel)
    if len(dels) == 2:
        print(min(dels), max(dels))
```

```python
for x in range(18782, 18822):
    dels = set()
    for delitel in range(2, int(x**0.5)+1):
        if x%delitel == 0:
            if delitel%2==1:
                dels.add(delitel)
            if (x//delitel)%2==1:
                dels.add(x//delitel)
    if len(dels) == 3:
        print(sorted(dels))
```
