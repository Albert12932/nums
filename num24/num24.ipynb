```python
s = open('24-d1.txt').read()

max_len = 0

current_len = 0

for element in s:
    if element == 'A':
        current_len += 1
    else:
        max_len = max(max_len, current_len)
        current_len = 0
    
print(max_len)
```

```python
s = open('24-d15.txt').readlines()

cnt = 0

for stroka in s:
    for i in range(1, len(stroka)-4):
        if stroka[i-1] == 'R' and stroka[i+1:i+5] == 'ZNOS':
            cnt += 1
            break
print(cnt)
```

```python
s = open('24var07.txt').read()

s = s.replace('12', '1 2').replace('21', '2 1').split(' ')

max_len = 0

for stroka in s:
    max_len = max(max_len, len(stroka))

print(max_len)
```

```python
s = open('24var16.txt').read()

current_len = 1
max_len = 1

for i in range(1, len(s)):
    if s[i] > s[i-1]:
        current_len += 1
    else:
        max_len = max(max_len, current_len)
        current_len = 1

print(max_len)
```
