# Логические операции

## <font color="#76c0ed">¬ A (отрицание)</font>        -  <font color = "#dfb639">not(A)</font>


## <font color="#76c0ed">A ^ B (умножение)</font>      -  <font color = "#dfb639">(A and B)</font>


## <font color="#76c0ed">A V B (сложение)</font>       -  <font color = "#dfb639">(A or B)</font>


## <font color="#76c0ed">A -> B (следование)</font>    -  <font color = "#dfb639">(A <= B)</font>


## <font color="#76c0ed">A = B (равносильность)</font> - <font color = "#dfb639">(A == B)</font>

```python
print('y z w x')
for x in 0,1:
    for y in 0,1:
        for z in 0,1:
            for w in 0,1:
                print(x,y,z,w)
```

```python
print('z x y')
for x in range(2):
    for y in range(2):
        for z in range(2):
            if not(   ((not(x)) or (not(z))) <= (x==y) ):
                print(z,x,y)
```

```python
print('w y z x')
for x in range(2):
    for y in range(2):
        for z in range(2):
            for w in range(2):
                if (  ((x and (not(y)) or (w <= z))) == (z==x)    ):
                    print(w,y,z,x)
```

```python
print('w x y z')
for x in range(2):
    for y in range(2):
        for z in range(2):
            for w in range(2):
                if ((not(y <= (x == w))) and (z <= x)):
                    print(w,x,y,z)
```
