```python
from turtle import *
tracer(0)
left(90)
m = 40
up()
back(10*m)
down()

for _ in range(7):
    forward(12*m)
    right(45)
    forward(6*m)
    right(135)

for x in range(-20, 20):
    for y in range(-20, 20):
        up()
        goto(x*m, y*m)
        down()
        dot(5, 'red')

update()
exitonclick()
```
