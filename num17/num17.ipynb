# 17 Номер ЕГЭ. Числовая последовательность

## Открытие файлов в Python

```python
nums = [int(i) for i in open('17-1.txt')]

average = sum(nums) / len(nums)

cnt = 0

max_sum = -10000000000

for i in range(len(nums)-1): 
    num_1 = nums[i]
    num_2 = nums[i+1]
    if num_1 > average and num_2 > average:
        if (num_1 + num_2) % 100 == 35:
            cnt += 1
            max_sum = max(max_sum, num_1+num_2)

print(cnt, max_sum)
```

```python
nums = [int(i) for i in open('17var01.txt.txt')]

cnt = 0
min_kratn = 1000000000
for x in nums:
    if x%21==0:
        min_kratn = min(min_kratn, x)
max_i = -1000000
for i in range(len(nums)-1):
    num_1 = nums[i]
    num_2 = nums[i+1]
    if num_1 % min_kratn == 0:
        cnt += 1
        max_i = max(max_i, num_1+num_2)

print(cnt, max_i)
```
