## <font color="ba922d">Тема 1: Булевы значения</font>

## Объяснение:
Булев тип (или логический тип) - это тип данных, который может принимать одно из двух возможных значений: True (истина) или False (ложь). В Python булевы значения используются для оценки условий и управления потоком выполнения программы.

## <font color="e85f35">Шаблон на русском:</font>
<font color="46e887"><pre>

</font>

## <font color="ba922d">Тема 2: Условия</font>

## Объяснение:

Условия в программировании позволяют выполнять определенный блок кода только в том случае, если определенное условие истинно.

## <font color="e85f35">Шаблон на русском:</font>

<pre><font color="46e887">если условие:
    выполнить действие

если другое условие:
    выполнить другое действие
    
иначе:
    выполнить третье действие</font></pre>


## <font color="46d5e8">Пример на Python:</font>

```python
x = int(input("Введите число"))
if x > 30:
    print("Число больше 30")
elif x > 20:
    print("Число больше 20")
elif x > 5:
    print("Число больше 5")
else:
    print("Число меньше 30")
```

```python
is_raining = True
is_sunny = False
if is_raining:
    print("Прихватите зонтик!")

if not is_sunny:
    print("Наденьте пальто!")
```

## <font color="ba922d">Тема 3: Циклы for</font>

## Объяснение:
Циклы for позволяют выполнять определенный блок кода для каждого элемента в последовательности (например, в списке).

## <font color="e85f35">Шаблон на русском:</font>
<pre>
<font color="46e887">for элемент in список:
    выполнить действие с элементом</pre></font>
<pre>
<font color="46e887">for элемент in промежуток(старт, конец, шаг):
    выполнить действие с элементом</pre></font>

## <font color="46d5e8">Пример на Python:</font>

```python
fruits = 'ABCDEF'
for fruit in fruits:
    print(fruit)
```

```python

for n in range():
    print(n**2)
```

## <font color="ba922d">Тема 4: циклы While</font>

## Объяснение:
Функция While является циклической конструкцией в Python, которая выполняет определенный блок кода, пока условие истинно. Этот тип цикла особенно удобен, когда заранее неизвестно, сколько раз нужно выполнить определенные действия.

## <font color="e85f35">Шаблон кода на русском:</font>
<pre>
<font color="46e887">while условие:
    блок кода
    выполняется, пока условие истинно</font></pre>


## <font color="46d5e8">Пример кода на языке Python:</font>


```python
count = 0
while count < 5:
    print(count)
    count += 1
```
